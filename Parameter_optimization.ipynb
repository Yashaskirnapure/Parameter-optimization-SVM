```python
from random import sample

def sample(df):
    s=df.sample(5000)
    y=s.iloc[:,16]
    X=s.drop(s.columns[16], axis=1)
    return X,y   
```

```python
X1,y1=sample(df)
X2,y2=sample(df)
X3,y3=sample(df)
X4,y4=sample(df)
X5,y5=sample(df)
X6,y6=sample(df)
X7,y7=sample(df)
X8,y8=sample(df)
X9,y9=sample(df)
X10,y10=sample(df)
```

```python
bestAccuracy=0
bestKernel=""
bestC=0
bestGamma=0
iteration=200
kernelList=['linear', 'poly', 'rbf', 'sigmoid']
```

```python
from sklearn.svm import SVC

def fitness(k,c,g,X,y):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=20)
    model=SVC(kernel = k, C = c,gamma=g).fit(X_train,y_train)
    y_pred = model.predict(X_test)  
    accuracy=accuracy_score(y_test,y_pred)*100
    return accuracy
    
```

```python
for i in range(iteration):
    k = np.random.choice(kernelList)
    c = np.random.uniform()
    g = np.random.uniform()
    accuracy = fitness(k,c,g,X1,y1)

    if accuracy > bestAccuracy:
        bestKernel = k
        bestC = c
        bestGamma = g
        best_accuracy = accuracy
        
```

CONVERGENCE OF SAMPLE 4

```python
values=[]
acc=[]
iteration=200
    
bestAccuracy=0
bestKernel=""
bestC=0
bestGamma=0

for i in range(iteration):
    k = np.random.choice(kernelList)
    c = np.random.uniform()
    g = np.random.uniform()
    accuracy = fitness(k,c,g,X4,y4)

    if accuracy > bestAccuracy:
        best_accuracy = accuracy
        values.append(i)
        acc.append(best_accuracy)   
    
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(values, acc)

ax.set_xlabel('Iteration')
ax.set_ylabel('Accuracy')

# Set the title of the plot
ax.set_title('Convergence Plot')

# Show the plot
plt.show()
```
